### Investigation overview
With the exploration done earlier, interesting finding were acquired. Among the findings that stood out were the difference between subscribers and customers in terms of bike usage and the number of rides they make.

### Dataset overview
The dataset has features about the individual rides made in the San Franscisco area.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
ford_df = pd.read_csv('FordGoBike.csv')
```

#### Visualization 1 (User type with the most rides)

```python
user_type_numbers = ford_df['user_type'].value_counts()

plt.pie(user_type_numbers, labels= user_type_numbers.index, startangle = 90, counterclock = False);
plt.title('Distribution of the user types')
plt.axis('square');
```

#### It is evident that subsribers are more than the customers

#### Visulization 2 (Geder with the most number of the riders)


```python
gender_of_riders = ford_df['member_gender'].value_counts() 

x = ['Male', 'Female', 'Other']
y = gender_of_riders
plt.bar(x, y, color = 'purple')
plt.xlabel('Gender of riders')
plt.ylabel('Number of rides')
plt.title('Number of rides against gender')
plt.show;
```

##### Males are the frequent riders

### Visualization 3 (distribution of the number of rides, user_type and gender of riders)

#### We plot a heatplot to visualize the count of the riders' genders together with their user type

```python
plt.figure(figsize=[14,6])
counts_of_gender_and_type = ford_df.groupby(['user_type','member_gender']).size()
counts_of_gender_and_type = counts_of_gender_and_type.reset_index(name = 'count')
counts_of_gender_and_type = counts_of_gender_and_type.pivot(index = 'user_type', columns = 'member_gender',values = 'count')
sns.heatmap(counts_of_gender_and_type, annot = True, fmt='d')
plt.title('Distribution of the number of rides, user_type and gender of riders¶');
```

##### Most of the riders comprise of male subscribers¶

#### Visualization 4 (relationship between member age, duration and user type)

#### Here, we shall plot a scatterplot to show how member age , duration and user type are inter-related.

```python
x = ford_df['member_age']
y = ford_df['duration_sec']
sns.scatterplot(x, y, hue=ford_df['user_type'], x_jitter=0.2)
plt.title('member age against duration and user type')
plt.xlabel('Member age (Years)')
plt.ylabel('Duration (Seconds)');
```

- In all cases younger riders between 20 and 45 years ride bikes for longer dureations.
- However, subscribers ride for longer periods of tme.

#### Visualization 5(Distribution of the ages of the riders)

###### We plot a histgram to show the distribution of the ages of the riders

```python
bin_edges = np.arange(20, ford_df['member_age'].max()+5, 5)
sns.histplot(ford_df['member_age'], bins = bin_edges, kde = False, color = 'green')
plt.title('Distribution of thr age of the riders')
plt.xlabel('Member age (Years)')
plt.ylabel('Frequency of rides');
```

##### Most riders are between 25 and 45 years
